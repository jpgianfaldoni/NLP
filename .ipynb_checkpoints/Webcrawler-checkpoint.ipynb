```python
from bs4 import BeautifulSoup
import urllib
from functools import partial
import mysql.connector
```

```python
request_url = urllib.request.urlopen('https://g1.globo.com/mundo/noticia/2022/02/23/china-acusa-eua-de-jogar-lenha-na-fogueira-na-crise-da-ucrania.ghtml') 
page = request_url.read().decode()
```

```python
soup = BeautifulSoup(page, 'html.parser')
```

```python
page_title = soup.title.string
page_text = soup.find_all('p')
page_string = ""
```

```python
for p in page_text:
    page_string += p.get_text()
print(page_string)
```

```python
all_links = soup.find_all('a')
for link in all_links:
    print(link.get('href'))
```

```python
def getPageSoup(url):
    request_url = urllib.request.urlopen(url)
    page = request_url.read().decode()
    soup = BeautifulSoup(page, 'html.parser')
    return soup
```

```python
def getPageText(soup):
    page_text = soup.find_all('p')
    page_string = ""
    for p in page_text:
        page_string += p.get_text()
    return page_string
```

```python
def getPageTitle(soup):
    return soup.title.string
```
